```python
Offence_Gender = Age_Offence_Data.groupby(["Offence","State","Gender"]).sum().reset_index()
Offence_Gender = Offence_Gender.drop(columns=['Age'])
Offence_Gender = Offence_Gender.pivot(index=["Offence","State"], columns='Gender', values='Victims').reset_index()

Age_Offence = Age_Offence_Data.groupby(["Offence","State","Age"]).sum().reset_index()
Age_Offence = Age_Offence.drop(columns=['Gender'])
Age_Offence = Age_Offence.pivot(index=["Offence","State"], columns='Age', values='Victims').reset_index()
Age_Offence
Offence_Gender
```
